```python
# def plot_prediction(pred, y):
#     plt.figure(figsize=(8, 6))
#     plt.scatter(X, y)
#     plt.scatter(X, pred)
#     plt.show()


def plot_prediction(X, pred, y):
    plt.figure(figsize=(8, 6))
    plt.scatter(X, y, color = 'red')
    # plt.scatter(X, pred)
    plt.plot(X, pred , color = 'green', linestyle = '--')
    plt.xlabel('hours') # X 축 이름
    plt.ylabel('score') # Y 축 이름
    plt.title('Score by hours') # 제목
    plt.show()
```

```python
# 훈련데이터 만들기
xy = (np.array([[1., 2., 3., 4., 5., 6.], [3., 6., 9., 12., 15., 18.]]))

X = xy[0].transpose()
Y = xy[1].transpose()

# X = np.random.randn(10)
# Y = 3 * X|

W = np.random.uniform(-1, 1)
b = np.random.uniform(-1, 1)

learning_rate = 0.01 

for epoch in range(10000):
    Y_Pred = W * X + b

    err = ((Y_Pred - Y)**2).mean()
    # print('epoch: {:10d}/1000, Error: {:10f}'.format(epoch, err))
          
    if err < 0.01 :
        break

    #경사하강법 적용 
    w_grad = learning_rate * ((Y_Pred - Y) * X).mean()
    b_grad = learning_rate * (Y_Pred - Y).mean()

    # W, b 값 갱신 
    W = W - w_grad
    b = b - b_grad

    if epoch % 50 == 0:
        print('epoch: {:10d}/1000, Error: {:10f}, Beta_gd: {:10f}, bias: {:10f}'.format(epoch, err, W.item(), b.item()))
        Y_Pred = W * X + b
        plot_prediction(X, Y_Pred, Y)
```
