```python
def plot_prediction(X, pred, y):
    plt.figure(figsize=(8, 6))
    plt.scatter(X, y, color = 'red')
    # plt.scatter(X, pred)
    plt.plot(X, pred , color = 'green', linestyle = '--')
    plt.xlabel('hours') # X 축 이름
    plt.ylabel('score') # Y 축 이름
    plt.title('Score by hours') # 제목
    plt.show()
```
