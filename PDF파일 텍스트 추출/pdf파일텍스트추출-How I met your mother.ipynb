```python
from pdfminer.high_level import extract_text
import time 
```

```python
path = '/home/ozzom/Workspace/Python/PDF파일 텍스트 추출/'
texts = extract_text(path+"how i met your mother 시즌1 e15-22.pdf")
```

```python
list_fullText = texts.split('How I met your mother Season 1 Episode ')   
```

```python
fileNm = 'How I met your mother Season 1 Episode '

for idx, episode in enumerate(list_fullText[1:]):
    list_text = episode.split('\n')
    
    with open (path + fileNm + str(idx+15) + '.txt', 'w', encoding='utf-8') as f:
        for text in list_text: 
            f.write("{}\n".format(text.strip()))
    time.sleep(1)
```
