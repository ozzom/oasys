```python
import cv2
import os
import time 
import ffmpeg 
```

```python
fileDictory = '/home/ozzom/Videos/LinearAlgebra'
resizedDirectory = '/home/ozzom/Videos/resized/'

files = os.listdir(fileDictory)
mp4s = [file for file in files if file[-3:] == 'mp4']
err = [] 

if not os.path.exists(fileDictory):
    os.mkdir(resizedDirectory)
    
for i, mp4 in enumerate(mp4s):
    filePath = fileDictory + '/' + mp4
    # print(filePath)
    try : 
        ffmpeg.input(filePath).output(resizedDirectory + mp4, crf=45, vsync='vfr').run()
        
#         # Progress Bar----------------------------------------------     
#         percentage = int((i + 1) * 100 / len(mp4s))
#         progress = percentage *'#' + (100 - percentage) * '-' 
#         print(f"\r{progress} {percentage}%", end="")    
#         time.sleep(1)
#         # ------------------------------------------------------------
        
    except:
        os.remove(resizedDirectory + mp4)
        err.append(resizedDirectory + mp4)            
```
