# 2. 축
### x 축, Y 축에 대한 Label 작성

```python
# 한글 깨짐 
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams['font.family'] = 'Malgun Gothic'  #Windows
matplotlib.rcParams['font.size'] = 15  # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False   # 한글폰트 사용시, 마이너스 글짜 깨짐 해결 
```

```python
x = [1, 2, 3]
y = [2, 4, 10]

plt.plot(x, y)
plt.title('꺾은선 그래프', fontdict={'family':'HYGungSo-Bold','size':20})  # 개별폰트 적용 
plt.xlabel('X축', color='red', loc='right', )    # X 축에 Label 입력 , 빨간색 적용, 글씨 위치 loc = left, center, right 
plt.ylabel('Y축',color='#00aa00',loc='top')    # Y 축에 Label 입력 , 녹색 적용 , 글씨 위치 loc = top, center, bottom 
```

```python
plt.plot(x, y)
plt.xticks([1, 2, 3])               # ticks를 이용해서 눈금자 조정 가능 
plt.yticks([3, 6, 9, 12])
plt.show()
```
