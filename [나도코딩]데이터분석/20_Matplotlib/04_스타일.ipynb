# 4.스타일
### 다양한 스타일 적용
### https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#


```python
# 한글 깨짐 
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams['font.family'] = 'Malgun Gothic'  #Windows
matplotlib.rcParams['font.size'] = 15  # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False   # 한글폰트 사용시, 마이너스 글짜 깨짐 해결 
```

```python
x = [1, 2, 3]
y = [2, 4, 10]
```

### 선 두께 조정 : linewidth = int
### 선 종류 : linestyle ='--','-', ':','-.', 'None'
### 데이터 부분 동그라미 표시 : marker = 'o', 'v'(삼각형), 'X', 'x' , marksize = int, markeredgecolor = 'red'
### 선 색깔 : color = 'b'(blue), 'g'(green), 'r'(red), 'c'(cyan), 'm'(magenta), 'y'(yellow), 'k'(black), 'w'(white)

```python
plt.plot(x, y, linewidth = 3, marker='o', markersize = 15, linestyle ='-.', color = 'k',  markeredgecolor = 'red')
```

## 포맷
### color, marker, linestyle

```python
plt.plot(x, y, 'ro--')  #  color, marker, linesytel 
```

```python
plt.plot(x, y, 'bv--')  #
```

```python
plt.plot(x, y, 'go:')
```

## 축약어 : 참조 https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html 
### markeredgecolor or mec, linestyle or ls, markersize or ms, markeredgecolor or mec 

```python
plt.plot(x, y, marker ='o', mfc='red', ms=10, mec = 'blue', ls =':')
```

## 라인 투명도 
### alpha : 0~1 사이의 값

```python
plt.plot(x, y, marker ='o', mfc='red', ms=10, mec = 'blue', ls =':',  alpha = 0.5)
```

## 그래프 크기 
## dpi는 확대율
## plt.figure(figsize(X비율, Y비율), dpi=0~1)))

```python
plt.figure(figsize =(10, 5))   # width, height 
plt.plot(x, y)
```

```python
plt.figure(figsize =(5, 10), dpi = 50)   # width, height , dpi(dots per inches ) 확대도
plt.plot(x, y)
```

## 그래프 배경색 
## fiuguar(facecolor='color')

```python
plt.figure(facecolor='yellow')
plt.plot(x,y)
```
