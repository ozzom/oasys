# 14.원그래프(심화)

### https://matplotlib.org/stable/gallery/pie_and_polar_charts/bar_of_pie.html

```python
import matplotlib.pyplot as plt 
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus']=False
```

```python
values = [30, 25, 20, 14, 10, 2]
label = ['Python', 'java', 'javascript','C#','C/C++', 'etc']
```

```python
# pie(value, label, autopct='%.1f%%'(숫자는 소수점, %%는 % 표시), 
#     startangle= 시작각도(0, 90, 180, 270), counterclock= 시계방향으로 돌 것인지(True,False) )
plt.pie(values, labels = label, autopct='%.1f%%', startangle=0, counterclock=False)   
plt.show()
```

# 색깔 조정

```python
# colors = ['b', 'g', 'r','c','m','y']
colors = ['#ffadad','#ffd6a5','#fdffb6','#caffbf','#9bf6ff','#a0c4ff']
explode =[0.05] * 6
plt.pie(values, labels=label, colors=colors, autopct='%.1f%%', startangle=0, counterclock=False, explode=explode)
plt.show()
```

# wedge pros
### 도넛처럼 가운데를 뚫게 만드는 기능

```python
wedgepros = {'width':0.8, 'edgecolor':'w', 'linewidth':2}  # 도넛처럼 가운데를 비우는 효과
plt.pie(values, labels=label, colors=colors, autopct='%.1f%%', startangle=90, counterclock=False, wedgeprops=wedgepros)
plt.show()
```

```python
# pct가 10 이상이면 표시, 이하면 표시 안함. 
def custom_autopct(pct): 
    # return ('%.1f%%' % pct) if pct> 4 else ''    아래와 같음 
    return '{:.1f}%'.format(pct) if pct > 4 else '' 

# pctdistance 가운데 쪽에 배치되도록 
plt.pie(values, labels=label, colors=colors, autopct=custom_autopct, startangle=90, counterclock=False, wedgeprops=wedgepros, pctdistance=0.7)
plt.show()
```
