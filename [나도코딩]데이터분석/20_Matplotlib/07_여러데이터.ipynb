## COVID-19 백신 종류별 접종 인구 

```python
# 아스트라제너카, 파이자, 모더나 접종자수, 단위 만명 
days = [1, 2, 3 ] # 1일, 2일, 3일 
az = [2, 4, 8]
pfizer= [4, 1, 3]
moderna = [1, 2, 5]

plt.plot(days, az)
plt.plot(days,pfizer)
plt.plot(days, moderna)
```

```python
plt.plot(days, az, label='astra zeneka' )
plt.plot(days,pfizer, label='pfzier', marker = 'o', linestyle='--')
plt.plot(days, moderna, label='moderna', marker = 'x', linestyle=':')

plt.legend()   # 범례 표시 
```

```python
plt.figure(figsize=(10, 5))
plt.plot(days, az, label='astra zeneka' )
plt.plot(days,pfizer, label='pfzier', marker = 'o', linestyle='--')
plt.plot(days, moderna, label='moderna', marker = '^', linestyle=':')

plt.legend(ncol = 3)   # 범례 표시      # 옆으로 표시 
```
