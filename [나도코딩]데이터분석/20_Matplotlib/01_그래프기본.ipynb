# Matplotlib
### 다양한 형태의 그래프를 통해서 데이터 시각화 할 수 있는 라이브러리

```python
import matplotlib.pyplot as plt
from random import * 
```

# 1. 그래프 기본
### X , Y 좌표 값을 이용한 그래프


```python
x = list()
y = list() 

for num in range(-10, 10):
    x.append(num)
    y.append(num*num+ 3*num)
    
plt.plot(x, y)
# plt.show
```

# 2. Title 설정

```python
plt.plot(x, y)
plt.title("1차원 그래프")
```
