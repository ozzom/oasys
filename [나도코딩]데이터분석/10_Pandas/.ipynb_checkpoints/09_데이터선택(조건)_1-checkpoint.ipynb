# 9. 데이터선택(조건) #1
조건에 해당하는 데이터 선택하기

```python
import pandas as pd

df= pd.read_excel('score.xlsx', index_col = '지원번호')
df
```

```python
df['키'] > 185 #  학생들의 키가 185 이상인지 여부를 True/False
```

```python
filt = (df['키'] >= 185)
df[filt]       # 조건을 만들어서 DataFrame에서 데이터를 선택적으로 가져올 수 있다. 
```

```python
df[~filt]   # 185 보다 작은 학생 들을 가져옴
```

```python
df[df['키'] >=185]
```

```python
df.loc[df['키'] >= 185, '수학'] # 키가 185 이상인 학생 들의 수학 점수 
```

```python
df.loc[df['키'] >= 185, ['이름', '과학', '과학']]
```

## 다양한 조건

#### & 그리고

```python
# 키 185 이상이고 북산고 학생인 
df.loc[(df['키']>=185) & (df['학교'] == '북산고')]
```

### 또는 |

```python
# 키는 170 미만 또는 키가 200 이상 
df.loc[ (df['키']<170) | (df['키'] > 200) ]
```

### str 함수  
>  str.startwith('str'), str.contains('str')

```python
# 송씨성을 가지고 있는 학생 조회 
filt = df['이름'].str.startswith('송')
df[filt]
```

```python
# 이름에 '태'가 들어가는 사람 
filt = df['이름'].str.contains('태')
df[filt]
```

```python
df[~filt]   # 이름에 '태' 라는 글자가 들어가는 사람 제외 
```

```python
langs = ['Python', 'Java']
filt = df['SW특기'].isin(langs)  #SW특기가 'Python' 이거나 'Java'인 값 가져오기 
df[filt]
```

```python
langs = ['python', 'java']
filt = df['SW특기'].str.lower().isin(langs)     # 대소문자 구분에 대한 문제 해결 방법, 모두 소문자로 바꿔서 비교 
df[filt]
```

```python
 # NaN  데이터가 포함되어 있으면, Contains로 접근하면 에러가 남, na=True or False 옵션으로 접근
filt = df['SW특기'].str.contains('Java', na=False) 
df[filt]
```
