## 데이터 제외하기 Drop
> df.dropna(inplace = True)

```python
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df.dropna(inplace=True)   # NAN  데이터를 삭제 
```

```python
df
```

## Dropna 데이터에 옵션넣기 
> dropna(axis= , how= )
- axis : index or columns  row(index)을 지울지, column을 지울 지 결정
- how : any or all  any (하나라도 있으면, ) all( 모두 일때 )

```python
df.dropna(axis='index', how='any')  # NaN이 하나라도 있는 index(row) 삭제 
```

```python
df.dropna(axis='columns', how='any')  # NaN이 하나라도 있는 column(열) 삭제  SW 특기가 사라짐 
```

```python
df['학교'] = np.nan
df
```

```python
df.dropna(axis='columns', how='all')  # 모든 칼럼 전체가 NaN인 경우에 삭제 
```
